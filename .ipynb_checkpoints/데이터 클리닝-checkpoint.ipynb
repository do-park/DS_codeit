# 정확성

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
df['abv'].quantile(0.25)
```

```python
df['abv'].quantile(0.75)
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
```

```python
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055
```

```python
df.loc[2250]
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace=True)
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df.plot(kind='box', y='abv')
```

# 영화 평점 분석하기

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/movie_metadata.csv')
```

```python
q1 = df['budget'].quantile(0.25)
q3 = df['budget'].quantile(0.75)
iqr = q3 - q1
```

```python
df['budget'] > q3 + 5 * iqr
```

```python
condition = (df['budget'] > q3 + 5 * iqr)
df.drop(df[condition].index, inplace=True)
```

```python
df.plot(kind='scatter', x='budget', y='imdb_score')
```

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/movie_metadata.csv')

drop_index = df['budget'].sort_values(ascending=False).head(15).index
df.drop(drop_index, inplace=True)
```

```python
df.plot(kind='scatter', x='budget', y='imdb_score')
```
